```python
import sys
sys.path.insert(0, "C:/Edu/VATSA")

import requests
from PIL import Image
from io import BytesIO
from modules.visual import VATSA_VisualPipeline

pipeline = VATSA_VisualPipeline(
    encoder_path="C:/Edu/VATSA/modules/visual/vatsa_visual_encoder_cifar10_deeper_unfreeze.pth"
)

# Download a real test image
url      = "https://ultralytics.com/images/bus.jpg"
response = requests.get(url)
img      = Image.open(BytesIO(response.content)).convert("RGB")
results  = pipeline.process_image(img)

print(f"Detections: {len(results)}")
for r in results:
    print(f"  {r['label']:<15} conf: {r['confidence']:.2f}  "
          f"embedding shape: {r['embedding'].shape}  "
          f"norm: {r['embedding'].norm():.4f}")
```
